```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Parameters
time_steps = 400  # Increase the number of steps for a longer time duration
time_duration = 20  # Total duration in seconds
time = np.linspace(0, time_duration, time_steps)  # x-axis values (time)

# Oscillating function with changing amplitude and added noise for sporadic behavior
amplitude = np.exp(-0.05 * time)  # Slower decay over a longer period
oscillating_function = amplitude * np.sin(2 * np.pi * time) + 0.1 * np.random.randn(time_steps)  # Adding noise

# Determine moving average window size for 3 seconds
time_interval = time[1] - time[0]  # Calculate the time step between data points
window_size = int(3 / time_interval)  # Number of points corresponding to 3 seconds

# Calculate moving average
moving_average = pd.Series(oscillating_function).rolling(window=window_size).mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, oscillating_function, label="Temporally fine estimation of H (new method)", color='blue')
plt.plot(time, moving_average, label="H Decoded using FT", color='orange')

# Labels and title
plt.xlabel("Time")
plt.ylabel("Hippocampal Gain")
plt.title("Methods of Decoding H")
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
